```python
import pandas as pd
```

```python
#In this step, I load all raw tables that were created in the previous notebook.
users = pd.read_csv("/Users/palakkakani/Desktop/ML/data/raw/users.csv")
user_activity = pd.read_csv("/Users/palakkakani/Desktop/ML/data/raw/user_activity.csv")
transactions = pd.read_csv("/Users/palakkakani/Desktop/ML/data/raw/transactions.csv")
experiment_assignment = pd.read_csv("/Users/palakkakani/Desktop/ML/data/raw/experiment_assignment.csv")
support_tickets = pd.read_csv("/Users/palakkakani/Desktop/ML/data/raw/support_tickets.csv")
```

```python
users.head()
```

```python
print(users.info())
```

```python
users['total_revenue'] = pd.to_numeric(users['total_revenue'], errors='coerce')
```

```python
users['total_revenue'].fillna(0, inplace=True)
```

```python
users['subscription_type'] = users['subscription_type'].astype(str)
```

```python
users['user_id'].duplicated().sum()
```

user activity

```python
user_activity.info()
```

```python
#Fill missing engagement values with 0
user_activity[['sessions_per_week','avg_watch_minutes','days_since_last_login']] = \
    user_activity[['sessions_per_week','avg_watch_minutes','days_since_last_login']].fillna(0)
```

```python
user_activity['device_type'] = user_activity['device_type'].astype(str)
user_activity['preferred_genre'] = user_activity['preferred_genre'].astype(str)
```

transaction

```python
transactions.info()
```

```python
## Fill missing financial indicators
transactions[['payment_success','failed_payment_count','chargeback_flag']] = \
    transactions[['payment_success','failed_payment_count','chargeback_flag']].fillna(0)
```

```python
transactions['monthly_revenue'] = pd.to_numeric(transactions['monthly_revenue'], errors='coerce').fillna(0)
```

experiment assignment

```python
experiment_assignment.info()
```

```python
experiment_assignment['experiment_group'] = experiment_assignment['experiment_group'].fillna('control')
```

```python
experiment_assignment['experiment_start_month'] = pd.to_datetime(
    experiment_assignment['experiment_start_month'], errors='coerce', format='%m'
)
# Convert month to datetime format
#This prepares the dataset for later A/B testing analysis.
```

support ticket

```python
support_tickets.info()
```

```python
support_tickets[['ticket_count','avg_response_time']] = \
    support_tickets[['ticket_count','avg_response_time']].fillna(0)
support_tickets['resolved_flag'] = support_tickets['resolved_flag'].fillna(0)
```

```python
# Ensure response time is within realistic range
support_tickets['avg_response_time'] = support_tickets['avg_response_time'].clip(1, 72)
```

```python
users.to_csv("/Users/palakkakani/Desktop/ML/data/processed/users.csv", index=False)
user_activity.to_csv("/Users/palakkakani/Desktop/ML/data/processed/user_activity.csv", index=False)
transactions.to_csv("/Users/palakkakani/Desktop/ML/data/processed/transactions.csv", index=False)
experiment_assignment.to_csv("/Users/palakkakani/Desktop/ML/data/processed/experiment_assignment.csv", index=False)
support_tickets.to_csv("/Users/palakkakani/Desktop/ML/data/processed/support_tickets.csv", index=False)
```
